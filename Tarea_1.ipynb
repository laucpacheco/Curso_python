3. Cree una estructura de datos que contenga información sobre un estudiante. La información requerida deberá ser el nombre, apellido paterno, apellido materno, rut, fecha de nacimiento (año- mes-día) y una nota (que representa su promedio). Luego, cree una lista de N estudiantes con datos aleatorios e imprima una vez la lista por cada uno de los siguientes órdenes:

  a. Orden alfabético apellido paterno, apellido materno
  b. Orden descendente del promedio
  c. Orden ascendente de las edades.

```python
import random

def crear_estudiante():
    nombres = ["Juan", "María", "Carlos", "Ana", "Luis", "Laura", "Pedro", "Mónica"]
    apellidos = ["Gómez", "López", "Martínez", "Pérez", "García", "Hernández", "Rodríguez"]

    estudiante = {}
    estudiante["nombre"] = random.choice(nombres)
    estudiante["apellido_paterno"] = random.choice(apellidos)
    estudiante["apellido_materno"] = random.choice(apellidos)
    estudiante["rut"] = f"{random.randint(10000000, 99999999)}-{random.randint(0, 9)}"
    estudiante["fecha_nacimiento"] = f"{random.randint(1990, 2020)}/{random.randint(1, 12)}/{random.randint(1, 31)}"
    estudiante["nota"] = round(random.uniform(1.0, 7.0))

    return estudiante

def lista_estudiantes(estudiantes, orden):
    print("Lista de estudiantes ordenados por", orden)

    if orden == "apellido":
        estudiantes_ordenados = sorted(estudiantes, key=lambda x: (x["apellido_paterno"], x["apellido_materno"]))
    elif orden == "promedio":
        estudiantes_ordenados = sorted(estudiantes, key=lambda x: x["nota"], reverse=True)
    elif orden == "edad":
        estudiantes_ordenados = sorted(estudiantes, key=lambda x: x["fecha_nacimiento"])
    else:
        print("Orden no válido.")
        return

    for estudiante in estudiantes_ordenados:
        print("Nombre:", estudiante["nombre"])
        print("Apellido Paterno:", estudiante["apellido_paterno"])
        print("Apellido Materno:", estudiante["apellido_materno"])
        print("RUT:", estudiante["rut"])
        print("Fecha de Nacimiento:", estudiante["fecha_nacimiento"])
        print("Nota:", estudiante["nota"])
        print("-" * 30)

N = 10
estudiantes = [crear_estudiante() for _ in range(N)]


lista_estudiantes(estudiantes, "apellido")
lista_estudiantes(estudiantes, "promedio")
lista_estudiantes(estudiantes, "edad")


```
